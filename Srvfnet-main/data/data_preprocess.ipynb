# Warped data process

```python
data = pd.read_csv("row_data/data_more_drastic.csv", header=None)
# delete NaNs
clean_data = data.dropna()
data = clean_data.values
data.shape
data_df = pd.DataFrame(data)
data_df
```

```python
num_data = data.shape[0]
np.random.seed(3407)
_new_order = list(range(num_data))
np.random.shuffle(_new_order)
data = data[_new_order, :]
data_df = pd.DataFrame(data)
data_df
```

```python
class_0_data_df = data_df[data_df.iloc[:, -1] == 0]
class_1_data_df = data_df[data_df.iloc[:, -1] == 1]
```

```python
print(f"In total we have {num_data} examples")
test_n = 4000
```

# True data

```python
data_true = pd.read_csv("row_data/allTrueSignals.csv", header=None)
# delete NaNs
clean_data_true = data_true.dropna()
new_column_data = [0.] * 3000 + [1.] * 3000
# 添加新列
clean_data_true['1000'] = new_column_data
data_true = clean_data_true.values
data_true = data_true[_new_order, :]
data_true_df = pd.DataFrame(data_true)
data_true_df
```

```python
class_0_true_data_df = data_true_df[data_true_df.iloc[:, -1] == 0]
class_1_true_data_df = data_true_df[data_true_df.iloc[:, -1] == 1]
```

```python
# keep test/valid

# train/valid
split = (4, 1)
num_tr_val = num_data - test_n
train_n = num_tr_val // sum(split) * split[0]
valid_n = num_tr_val // sum(split) * split[1]

class_0_true_data_df = class_0_true_data_df.values
class_1_true_data_df = class_1_true_data_df.values

# for test and valid, randomly choose test/2 and valid_n/2 from each class to build a balance test/valid set (the rest examples are left for the train set)
n_test_class_0_split = test_n // 2
n_test_class_1_split = test_n - n_test_class_0_split

n_valid_class_0_split = valid_n // 2
n_valid_class_1_split = valid_n - n_valid_class_0_split

raw_data_true_test_0 = pd.DataFrame(class_0_true_data_df[:n_test_class_0_split, :])
raw_data_true_valid_0 = pd.DataFrame(class_0_true_data_df[n_test_class_0_split:(n_test_class_0_split+n_valid_class_0_split), :])
raw_data_true_train_0 = pd.DataFrame(class_0_true_data_df[(n_test_class_0_split+n_valid_class_0_split):, :])

raw_data_true_test_1 = pd.DataFrame(class_1_true_data_df[:n_test_class_1_split, :])
raw_data_true_valid_1 = pd.DataFrame(class_1_true_data_df[n_test_class_1_split:(n_test_class_1_split+n_valid_class_1_split), :])
raw_data_true_train_1 = pd.DataFrame(class_1_true_data_df[(n_test_class_1_split+n_valid_class_1_split):, :])

# merge
raw_data_true_train = pd.concat([raw_data_true_train_0, raw_data_true_train_1])
raw_data_true_valid = pd.concat([raw_data_true_valid_0, raw_data_true_valid_1])
raw_data_true_test = pd.concat([raw_data_true_test_0, raw_data_true_test_1])

len(raw_data_true_train), len(raw_data_true_valid), len(raw_data_true_test)
```

```python
d_len = data_true_df.shape[1] - 1 # last col
raw_data_true_train[d_len] = raw_data_true_train[d_len].astype(int)
raw_data_true_valid[d_len] = raw_data_true_valid[d_len].astype(int)
raw_data_true_test[d_len] = raw_data_true_test[d_len].astype(int)
```

```python
raw_data_true_train.to_csv('train_true.csv', index=False, header=False)
raw_data_true_valid.to_csv('valid_true.csv', index=False, header=False)
raw_data_true_test.to_csv('test_true.csv', index=False, header=False)
```
